```python
import os
import numpy as np
import pandas as pd

def add_ticker_column(df, filename, index):
    """Adds a ticker column to the DataFrame based on the filename."""
    ticker = filename.split('.')[0]  # Extract ticker from filename
    df['Ticker'] = ticker
    df['Index'] = index
    return df

def load_and_combine_csv_data(base_dir, ticker_df):
    """
    Loads CSV files only for tickers present in ticker_df, adds a Ticker column,
    and combines them into a single DataFrame.
    """
    if ticker_df is None:
        print("Error: ticker_df is None.  Ensure it's loaded before calling this function.")
        return None

    valid_tickers = ticker_df['Ticker'].str.upper().tolist()  # Get list of valid tickers, convert to uppercase

    all_data = []
    for subdir in os.listdir(base_dir): # go through base, where stock market dir is
        subdir_path = os.path.join(base_dir, subdir)
        if os.path.isdir(subdir_path): # If is dir

            csv_dir = os.path.join(subdir_path, "csv") # go to the csv dir inside the base
            if os.path.exists(csv_dir):
                for filename in os.listdir(csv_dir):
                    if filename.endswith(".csv"):
                        ticker = filename.split('.')[0].upper()  # Extract ticker and convert to uppercase

                        if ticker in valid_tickers:  # Check if ticker is in the valid list
                            csv_path = os.path.join(csv_dir, filename)
                            try:
                                df = pd.read_csv(csv_path, parse_dates=['Date'])  # Parse date

                                # Date is not defined, print so we know
                                if 'Date' not in df:
                                    print(f"Issue, Date column not defined in {filename}")

                                df = add_ticker_column(df, filename, subdir_path)  # ticker to df from file name
                                all_data.append(df)  # Append the df to all_data

                            except Exception as e:
                                print(f"Error loading or processing {filename}: {e}")
                        else:
                            print(f"Skipping {filename}: Ticker not found in ticker_df.")

            else:
                print(f"csv directory not found in {subdir}")

    if not all_data:
        print("No data loaded. Ensure the base directory contains the specified tickers, and the CSV files are correctly named.")
        return None

    combined_df = pd.concat(all_data, ignore_index=True)  # combine
    return combined_df

# --- Main ---
base_dir = "/content/stock_market_data/"  # The main directory with the four folders

# ASSUMPTION: ticker_df is already loaded and accessible in this scope!

combined_stock_data = load_and_combine_csv_data(base_dir, ticker_df)  # Loads data for tickers present in ticker_df.  Pass ticker_df to the function

if combined_stock_data is not None:
    print(f"Combined data shape: {combined_stock_data.shape}")
    print(f"Combined data columns: {combined_stock_data.columns}")
    print(combined_stock_data.head())  # Print the head of the data
else:
    print("Failed to load combined stock data.")
```

```python
import pandas as pd
import time
import numpy as np  # Import numpy

def calculate_market_changes(df):
    """
    Calculates 'Regular_Market_Change' and 'After_Hours_Market_Change' for a DataFrame.

    Filters out rows with 0.0 in 'Open', 'Close', or 'Next_Day_Open'.

    Args:
        df (pd.DataFrame): DataFrame with columns 'Open', 'Close', 'Next_Day_Open'.

    Returns:
        pd.DataFrame: DataFrame with added columns 'Regular_Market_Change' and 'After_Hours_Market_Change',
                      or an empty DataFrame if required columns are missing.  Returns a copy to avoid SettingWithCopyWarning
    """
    required_columns = ['Open', 'Close', 'Next_Day_Open', 'Ticker', 'Date']  # Added Ticker and Date since they're used in later steps
    if not all(col in df.columns for col in required_columns):
        missing_cols = [col for col in required_columns if col not in df.columns]
        print(f"Error: DataFrame must contain the following columns: {missing_cols}")
        return pd.DataFrame()

    # Filter out rows with 0.0 in Open, Close, or Next_Day_Open
    df_filtered = df[
        (df['Open'] != 0.0) &
        (df['Close'] != 0.0) &
        (df['Next_Day_Open'] != 0.0)
    ].copy()

    # Calculate Market Changes (after filtering)
    df_filtered['Regular_Market_Change'] = ((df_filtered['Close'] - df_filtered['Open']) / df_filtered['Open']) * 100
    df_filtered['After_Hours_Market_Change'] = ((df_filtered['Next_Day_Open'] - df_filtered['Close']) / df_filtered['Close']) * 100

    return df_filtered


def get_market_changes(df):
    """
    Calculates market changes and returns all rows with calculated changes.

    Args:
        df (pd.DataFrame): DataFrame containing stock data.

    Returns:
        tuple: A tuple containing two DataFrames:
               - DataFrame with 'Regular_Market_Change'.
               - DataFrame with 'After_Hours_Market_Change'.
               Returns (None, None) if market changes could not be calculated due to missing columns.
    """
    df_with_changes = calculate_market_changes(df)

    if df_with_changes.empty:
        print("Could not calculate market changes due to missing data or errors.")
        return None, None

    regular_changes = df_with_changes.copy()  # all with regular changes
    after_hours_changes = df_with_changes.copy()  # all with after hours changes

    return regular_changes, after_hours_changes

def convert_date_to_unix(df, date_column='Date'):
    """Converts the 'Date' column in a DataFrame to Unix timestamps.

    Args:
        df (pd.DataFrame): The DataFrame to modify.
        date_column (str): The name of the column containing the date.  Defaults to 'Date'.

    Returns:
        pd.DataFrame: The modified DataFrame with the 'Date' column converted to Unix timestamps,
                      or None if the 'Date' column is not found.
    """
    if date_column not in df.columns:
        print(f"Error: '{date_column}' column not found in DataFrame.")
        return None

    # Convert the 'Date' column to datetime objects, handling errors
    try:
        df['Date_Timestamp'] = pd.to_datetime(df[date_column], errors='raise').astype('int64') // 10**9  # Convert to seconds
        #This creates a new column called 'Date_Timestamp' containing unix timestamps
    except ValueError as e:
        print(f"Error converting '{date_column}' to datetime: {e}")
        return None

    return df

# Convert Date to Unix timestamp
final_stock_data = convert_date_to_unix(final_stock_data)

if final_stock_data is not None:
    print("DataFrame with Unix timestamps:")
    print(final_stock_data[['Date', 'Date_Timestamp']].head()) # Print only the Date and timestamp columns for brevity

    regular_changes, after_hours_changes = get_market_changes(final_stock_data)

    if regular_changes is not None and after_hours_changes is not None:
        # Sort the DataFrames
        top_regular = regular_changes[regular_changes['Regular_Market_Change'] > 0].sort_values(by='Regular_Market_Change', ascending=False).copy()
        top_after_hours = after_hours_changes[after_hours_changes['After_Hours_Market_Change'] > 0].sort_values(by='After_Hours_Market_Change', ascending=False).copy()

        print("\nAll Regular Market Changes (Positive and Sorted):")
        print(top_regular)

        print("\nAll After Hours Market Changes (Positive and Sorted):")
        print(top_after_hours)
    else:
        print("Failed to calculate and retrieve market changes.")
else:
    print("Failed to convert Date to Unix timestamp.")
```

```python

#Working

from gnews import GNews
from datetime import datetime, timedelta
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random  # For introducing random delays
from tqdm import tqdm  # For visualizing long loading
from newspaper import Article # Ensure newspaper is present


def get_news_from_gnews(ticker, date, max_retries=3, delay=1,gnews_language='en',gnews_country='US'):
    """
    Fetches news articles from GNews, with retries and rate limiting.

    Args:
        ticker: The stock ticker symbol.
        date: The date to search for news (as a datetime object).
        max_retries: Maximum number of retries for failed requests.
        delay: Delay in seconds between retries.

    Returns:
        A list of news article dictionaries (title, url, content).
    """
    google_news = GNews(language=gnews_language, country=gnews_country)
    date_str = date.strftime('%Y-%m-%d')
    query = f"{ticker} stock"
    articles = []

    for attempt in range(max_retries):
        try:
            news_urls = google_news.get_news(query) # try getting the articles
            if news_urls is not None:
                for news in news_urls: # now go through the urls

                    try:
                        url = news['url'] #Set the URL to be the URL of the article.

                        article = get_article_content(url,max_retries=max_retries, delay=delay) # this will run our helper for article retrieval.

                        if article:
                            articles.append({
                                "title": article.title if article else "Title not found",
                                "url": url,
                                "content": article.text if article else "Content not found",
                            })
                        else:
                            print(f"Could not retrieve full article from {url}")
                    except Exception as e:
                        print(f"Error retrieving article from {url}: {e}")
            else:
                print('news_urls is none')
            # Introduce a delay to avoid rate limiting
            time.sleep(delay + random.uniform(0, 3))  # Add a small random delay

            if articles:  # If we got any articles, return
                return articles

        except Exception as e:
            print(f"Attempt {attempt + 1} failed to fetch news for {ticker} on {date}: {e}")
            if attempt < max_retries - 1:
                time.sleep(delay)  # Wait before retrying
            else:
                print("Max retries reached. Giving up.")
                return []

    return []  # Return empty list if all retries fail


def get_article_content(url, max_retries = 3, delay = 5):
     for attempt in range(max_retries):
        try:

            google_news = GNews()
            article = google_news.get_full_article(url)
            if article:
                return article
            else:
                return None
        except Exception as e:
            print(f"Attempt {attempt + 1} article retrieval for {url} failed, retrying {e}")
            if attempt < max_retries - 1:
                time.sleep(delay)  # Wait before retrying
            else:
                print("Max retries reached. Giving up retrieval of article.")
                return None

def get_top_3_articles(related_news):
    """
    Extracts the top 3 article titles by visiting each URL and scraping the title from the HTML,
    handling React Helmet titles.

    Args:
        related_news: A list of dictionaries containing news article data (with 'url' key).

    Returns:
        A list of the titles of the top 3 articles, fetched from the actual web pages.
    """
    if not isinstance(related_news, list):
        print("Error: related_news is not a list.")
        return []

    top_3_titles = []
    for i, news_item in enumerate(related_news):
        if i >= 3:
            break  # only retrieves top 3

        if isinstance(news_item, dict) and 'url' in news_item:
            url = news_item['url']
            try:
                response = requests.get(url, timeout=5)  # Add timeout
                response.raise_for_status()  # Raise HTTPError for bad responses (4xx or 5xx)

                soup = BeautifulSoup(response.content, 'html.parser')

                # Target the title tag with the data-react-helmet attribute
                title_tag = soup.find('title', {'data-react-helmet': 'true'})
                if title_tag:
                    title = title_tag.text
                else:
                    title = url # set a standard to return no title

                top_3_titles.append(title)
            except requests.exceptions.RequestException as e:
                print(f"Error fetching URL {url}: {e}")
                top_3_titles.append(f"Error: Could not fetch title from URL (Network Error)")
            except AttributeError as e:
                print(f"Title not found in the article. Returning content")
                if isinstance(news_item,dict) and 'title' in news_item:
                    top_3_titles.append(news_item['title'])
            except Exception as e:
                print(f"Error processing URL {url}: {e}")
                top_3_titles.append(f"Error: Could not extract title from URL (HTML Parsing Error)")


        else:
            print(f"Skipping: Invalid format for news_item number {i+1}")

    return top_3_titles

tqdm.pandas(desc="Fetching News")  # Initialize tqdm for pandas.
related_news = top_50_after_hours.progress_apply(
    lambda row: get_news_from_gnews(row['Ticker'], row['Date']),
    axis=1
)

#Now here is the important change. apply and set
top_50_after_hours['top_3_articles'] = related_news.apply(get_top_3_articles)

# Display the DataFrame with the related news:
print(top_50_after_hours[['Date', 'Ticker', 'top_3_articles']].head())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import requests
import json

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")

# Initialize WebDriver
driver = webdriver.Chrome(options=chrome_options)

def extract_title_from_url_Selenium(url):
    """
    Extracts the title and final URL from a dynamically rendered web page using Selenium and BeautifulSoup.

    Args:
        url: The URL of the web page.

    Returns:
        A tuple containing the title and the final URL, or (None, None) if extraction fails.
    """
    # Extract title from the best available source
    title = None
    final_url = None

    try:
        # Open the Google News RSS link
        driver.get(url)

        # Wait for the page to load (increased to 5 seconds)
        time.sleep(5)

        # Get the final redirected URL
        final_url = driver.current_url

        # Get page source and parse it
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")

        # 1. Try extracting from <meta property="og:title">
        meta_og_title = soup.find("meta", property="og:title")
        if meta_og_title and meta_og_title.get("content"):
            title = meta_og_title["content"]

        # 2. If not found, try <meta name="title">
        if not title:
            meta_name_title = soup.find("meta", attrs={"name": "title"})
            if meta_name_title and meta_name_title.get("content"):
                title = meta_name_title["content"]

        # 3. If still not found, fall back to <title>
        if not title:
            title_tag = soup.find("title")
            if title_tag:
                title = title_tag.text.strip()

        # If no title was found at all
        if not title:
            title = "Title not found"
        return (title, final_url)

    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL {url}: {e}")
        return (f"Error: Could not fetch title from URL (Network Error): {e}",url)

    except Exception as e:
        print(f"Error parsing URL {url}: {e}")
        return (f"Error: Could not extract title from URL (HTML Parsing Error): {e}",url)

    finally:
        return(title, final_url)

def get_title_and_url_string(urls):
    """
    Takes a comma-separated list of URLs, extracts the title and final URL for each,
    and returns a concatenated string of 'title: X, url: Y' pairs.

    Args:
        urls: A comma-separated string of URLs.

    Returns:
        A string containing the title and URL for each article, concatenated together.
    """
    list_urls = urls  # Splits the list to smaller items
    result_string = ""

    for item in list_urls:
        url = item.strip()
        title, final_url = extract_title_from_url_Selenium(url)  # extract title and URL

        # Add the 'title: X, url: Y' pair to the string
        result_string += f"title: {title}, url: {final_url}; "

    return result_string
```

```python
import json
import pandas as pd

def load_json_to_dataframe(filename, chunksize=10000):
    """Loads a JSON file into a Pandas DataFrame in chunks to handle large files.

    Args:
        filename (str): The path to the JSON file.
        chunksize (int): The number of lines to read into the DataFrame at a time.

    Returns:
        pandas.DataFrame: The resulting DataFrame, or None if there was an error.
    """
    try:
        data = []
        with open(filename, 'r', encoding='utf-8') as f:
            while True:
                lines = []
                for _ in range(chunksize):
                    try:
                        line = next(f).strip()
                        if line:  # Check if line is not empty
                            try:
                                json_object = json.loads(line)
                                lines.append(json_object)
                            except json.JSONDecodeError as e:
                                print(f"JSONDecodeError: {e} for line: {line}")
                                # Potentially log the bad line to a file for later inspection
                    except StopIteration:
                        break  # End of file
                if not lines:
                    break #No more data

                df_chunk = pd.DataFrame(lines)
                data.append(df_chunk)

        if data:  #Check if any data was loaded
            df = pd.concat(data, ignore_index=True)
            print(f"Loaded data into DataFrame with shape: {df.shape}")
            return df
        else:
            print("No data loaded.")
            return None


    except FileNotFoundError:
        print(f"Error: File not found: {filename}")
        return None
    except Exception as e:
        print(f"An error occurred while loading the JSON data: {e}")
        return None

# Example Usage:
filename = "TickerTick-stock-news-dataset.2023-11-23.json"  # Replace with your actual filename

df = load_json_to_dataframe(filename)

if df is not None:
    # Now you can work with the DataFrame 'df'
    print(df.head())  # Print the first few rows
    # Example: df.describe() to get summary statistics
else:
    print("Failed to load data into DataFrame.")

print("Done!")
```

```python
import pandas as pd

def query_news_by_date_and_ticker(df, date_str, ticker_str):
    """Queries the news DataFrame for entries matching a date and ticker.

    Args:
        df (pd.DataFrame): The news DataFrame with a 'Date' column (YYYY-MM-DD)
                            and 'tickers_direct' and 'tickers_indirect' columns.
        date_str (str): The date string in YYYY-MM-DD format.
        ticker_str (str): The ticker string to match.

    Returns:
        pandas.DataFrame: A DataFrame containing the matching news entries, or an empty
                      DataFrame if no matches are found.  Returns None if an error occurred.
    """
    if 'Date' not in df.columns or 'tickers_direct' not in df.columns or 'tickers_indirect' not in df.columns:
        print("Error: DataFrame must contain 'Date', 'tickers_direct', and 'tickers_indirect' columns.")
        return None

    try:
        # Convert ticker string to lowercase
        ticker_str_lower = ticker_str.lower()

        # **First filter by date**
        date_filtered_df = df[df['Date'] == date_str].copy()

        # Check if the date filter resulted in an empty DataFrame
        if date_filtered_df.empty:
            print(f"No news found for Date '{date_str}'.")
            return date_filtered_df # Returns empty dataframe, not None

        # Create a boolean mask for matching the ticker in tickers_direct
        # Check if the column exists
        if 'tickers_direct' in date_filtered_df.columns:
            if pd.api.types.is_string_dtype(date_filtered_df['tickers_direct']):
                tickers_direct_mask = date_filtered_df['tickers_direct'].str.lower().str.contains(ticker_str_lower, na=False)
            else:
                 # Attempt conversion to string dtype
                try:
                    tickers_direct_mask = date_filtered_df['tickers_direct'].astype(str).str.lower().str.contains(ticker_str_lower, na=False)
                except Exception as e:
                    print(f"Could not convert tickers_direct to string: {e}")
                    tickers_direct_mask = pd.Series([False] * len(date_filtered_df), index=date_filtered_df.index)

        else:
            tickers_direct_mask = pd.Series([False] * len(date_filtered_df), index=date_filtered_df.index)  # No match if column doesn't exist


        # Create a boolean mask for matching the ticker in tickers_indirect
        # Check if the column exists
        if 'tickers_indirect' in date_filtered_df.columns:
            if pd.api.types.is_string_dtype(date_filtered_df['tickers_indirect']):
                tickers_indirect_mask = date_filtered_df['tickers_indirect'].str.lower().str.contains(ticker_str_lower, na=False)
            else:
                 # Attempt conversion to string dtype
                try:
                    tickers_indirect_mask = date_filtered_df['tickers_indirect'].astype(str).str.lower().str.contains(ticker_str_lower, na=False)
                except Exception as e:
                    print(f"Could not convert tickers_indirect to string: {e}")
                    tickers_indirect_mask = pd.Series([False] * len(date_filtered_df), index=date_filtered_df.index)

        else:
            tickers_indirect_mask = pd.Series([False] * len(date_filtered_df), index=date_filtered_df.index)  # No match if column doesn't exist

        # Combine the masks using logical OR
        combined_mask = (tickers_direct_mask | tickers_indirect_mask)

        # Filter the DataFrame using the combined mask
        matching_news = date_filtered_df[combined_mask].copy()

        return matching_news

    except Exception as e:
        print(f"Error during query: {e}")
        return None  # Consistent return of None for errors


#  Matching Loop and Adding to DataFrames

if 'top_regular' in locals() and 'top_after_hours' in locals() and 'news_df' in locals():

    # Add 'matched_news' column to top_regular and top_after_hours (initialize as empty lists)
    top_regular['matched_news'] = [[] for _ in range(len(top_regular))]
    top_after_hours['matched_news'] = [[] for _ in range(len(top_after_hours))]


    # Loop through top_regular
    print("Matching news with top_regular:")
    for index, row in top_regular.iterrows():
        # Check if 'Date' is datetime or string
        if isinstance(row['Date'], str):
            date_str = row['Date'][:10]  # Extract YYYY-MM-DD from Date string
        else:  # Assuming it's a datetime object
            date_str = row['Date'].strftime('%Y-%m-%d')  # Format as YYYY-MM-DD

        ticker_str = row['Ticker']  # Get the ticker

        matching_news = query_news_by_date_and_ticker(news_df, date_str, ticker_str)

        if matching_news is not None:
            if not matching_news.empty:
                print(f"  Match found for Date: {date_str}, Ticker: {ticker_str}")
                # Add the matching news to the 'matched_news' column
                top_regular.at[index, 'matched_news'] = matching_news.to_dict('records')
            else:
                print(f"  No news found for Date: {date_str}, Ticker: {ticker_str}")
        else:
            print(f"  Error during query for Date: {date_str}, Ticker: {ticker_str}")


    # Loop through top_after_hours
    print("\nMatching news with top_after_hours:")
    for index, row in top_after_hours.iterrows():
        # Check if 'Date' is datetime or string
        if isinstance(row['Date'], str):
            date_str = row['Date'][:10]  # Extract YYYY-MM-DD from Date string
        else:  # Assuming it's a datetime object
            date_str = row['Date'].strftime('%Y-%m-%d')  # Format as YYYY-MM-DD

        ticker_str = row['Ticker']  # Get the ticker

        matching_news = query_news_by_date_and_ticker(news_df, date_str, ticker_str)

        if matching_news is not None:
            if not matching_news.empty:
                print(f"  Match found for Date: {date_str}, Ticker: {ticker_str}")
                # Add the matching news to the 'matched_news' column
                top_after_hours.at[index, 'matched_news'] = matching_news.to_dict('records')
            else:
                print(f"  No news found for Date: {date_str}, Ticker: {ticker_str}")
        else:
            print(f"  Error during query for Date: {date_str}, Ticker: {ticker_str}")

    # Print the modified DataFrames (or just the relevant columns)
    print("\nModified top_regular with matched_news:")
    print(top_regular[['Date', 'Ticker', 'matched_news']].head())  # Show only relevant columns

    print("\nModified top_after_hours with matched_news:")
    print(top_after_hours[['Date', 'Ticker', 'matched_news']].head())  # Show only relevant columns

else:
    print("Error: top_regular, top_after_hours, or news_df not found. Load DataFrames before running")
```

```python
from jupyter_dash import JupyterDash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import dash

# Clean your DataFrame
top_after_hours_df = top_after_hours_df.sort_values(by='Date')
top_after_hours_df = top_after_hours_df.drop_duplicates(subset=['Ticker', 'Date'], keep='first')
top_after_hours_df = top_after_hours_df.dropna(subset=['After_Hours_Market_Change'])

# Create app
app = JupyterDash(__name__)
app.title = "After-Hours Movers Dashboard"

# Layout
app.layout = html.Div([
    html.H1("Top After-Hours Movers", style={"textAlign": "center"}),

    # Date filter
    dcc.DatePickerRange(
        id='date-range',
        min_date_allowed=top_after_hours_df['Date'].min(),
        max_date_allowed=top_after_hours_df['Date'].max(),
        start_date=top_after_hours_df['Date'].min(),
        end_date=top_after_hours_df['Date'].max()
    ),

    html.Br(), html.Br(),

    # Sector filter
    html.Div([
        html.Label("Filter by Sector:"),
        dcc.Dropdown(
            id='sector-dropdown',
            options=[{'label': s, 'value': s} for s in sorted(top_after_hours_df['Sector'].dropna().unique())],
            value=None,
            placeholder="All Sectors",
            clearable=True,
            style={'width': '300px'}
        )
    ]),

    html.Br(),

    # Top N input with suggestions
    html.Div([
        html.Label("Select Top N Tickers:", style={"marginRight": "10px"}),

        dcc.Input(
            id='top-n',
            type='number',
            min=1,
            max=len(top_after_hours_df['Ticker'].unique()),
            value=10,
            debounce=True,
            style={'width': '100px'}
        ),

        html.Div([
            html.Button('Top 10', id='top-10-btn', n_clicks=0, style={'marginRight': '5px'}),
            html.Button('Top 25', id='top-25-btn', n_clicks=0, style={'marginRight': '5px'}),
            html.Button('Top 50', id='top-50-btn', n_clicks=0)
        ], style={'marginTop': '10px'})
    ]),

    html.Br(),

    # Bar chart
    dcc.Graph(id='bar-top-movers'),

    # Hover info
    html.Div([
        html.H4("Details"),
        html.Div(id='bar-hover-info', style={
            'border': '1px solid #ddd',
            'padding': '10px',
            'backgroundColor': '#f9f9f9',
            'marginTop': '10px'
        })
    ]),

    # Ticker inspection
    html.H3("Inspect Ticker"),
    dcc.Dropdown(id='ticker-dropdown', placeholder="Select a ticker..."),
    dcc.Graph(id='line-chart'),
])

# Callback: quick buttons to change top N
@app.callback(
    Output('top-n', 'value'),
    Input('top-10-btn', 'n_clicks'),
    Input('top-25-btn', 'n_clicks'),
    Input('top-50-btn', 'n_clicks'),
    prevent_initial_call=True
)
def set_top_n(n10, n25, n50):
    ctx = dash.callback_context
    if not ctx.triggered:
        return dash.no_update
    button_id = ctx.triggered[0]['prop_id'].split('.')[0]
    if button_id == 'top-10-btn': return 10
    if button_id == 'top-25-btn': return 25
    if button_id == 'top-50-btn': return 50
    return dash.no_update

# Callback: update bar chart + dropdown
@app.callback(
    Output('bar-top-movers', 'figure'),
    Output('ticker-dropdown', 'options'),
    Input('date-range', 'start_date'),
    Input('date-range', 'end_date'),
    Input('top-n', 'value'),
    Input('sector-dropdown', 'value')
)
def update_bar_chart(start_date, end_date, top_n, selected_sector):
    if not top_n:
        top_n = 10

    filtered = top_after_hours_df[
        (top_after_hours_df['Date'] >= start_date) &
        (top_after_hours_df['Date'] <= end_date)
    ]

    # Apply sector filter
    if selected_sector:
        filtered = filtered[filtered['Sector'] == selected_sector]

    # Cap extreme outliers (optional threshold)
    filtered = filtered[filtered['After_Hours_Market_Change'] < 100]

    if filtered.empty:
        return px.bar(title="No data in selected range"), []

    top_movers = filtered.sort_values(by='After_Hours_Market_Change', ascending=False).head(top_n)

    fig = px.bar(
        top_movers,
        x='Ticker',
        y='After_Hours_Market_Change',
        color='Sector',
        title="Top After-Hours Movers",
        hover_data={
            'Ticker': True,
            'Sector': True,
            'After_Hours_Market_Change': ':.2f',
            'Open': ':.2f',
            'High': ':.2f',
            'Low': ':.2f',
            'Close': ':.2f',
            'Volume': ':,',
            'Date': True
        }
    )

    fig.update_layout(yaxis_tickformat=".2f")

    options = [{'label': t, 'value': t} for t in top_movers['Ticker'].unique()]
    return fig, options

# Callback: bar hover info panel
@app.callback(
    Output('bar-hover-info', 'children'),
    Input('bar-top-movers', 'hoverData')
)
def display_hover_info(hoverData):
    if not hoverData:
        return "Hover over a bar to see details."

    ticker = hoverData['points'][0]['x']
    row = top_after_hours_df[top_after_hours_df['Ticker'] == ticker].sort_values('Date', ascending=False).head(1).iloc[0]

    return html.Div([
        html.P(f"Ticker: {row['Ticker']}"),
        html.P(f"Sector: {row['Sector']}"),
        html.P(f"Date: {row['Date'].date()}"),
        html.P(f"Open: {row['Open']:.2f}"),
        html.P(f"High: {row['High']:.2f}"),
        html.P(f"Low: {row['Low']:.2f}"),
        html.P(f"Close: {row['Close']:.2f}"),
        html.P(f"Volume: {int(row['Volume']):,}"),
        html.P(f"After-Hours % Change: {row['After_Hours_Market_Change']:.2f}%")
    ])

# Callback: line chart
@app.callback(
    Output('line-chart', 'figure'),
    Input('ticker-dropdown', 'value')
)
def update_line_chart(ticker):
    if not ticker:
        return px.line(title="Select a ticker to see Close Price Trend")

    df = top_after_hours_df[top_after_hours_df['Ticker'] == ticker]
    fig = px.line(df, x='Date', y='Close', title=f"{ticker} Close Price Over Time")
    return fig

# Run in Google Colab
app.run_server(mode='inline', debug=True)
```
